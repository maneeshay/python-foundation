# Topic 1: Variables & Python Object Model

## 1. Core Principle

In Python, a variable is a name bound to an object.

Variables do not store values.
They reference objects that live in memory.

```python
x = 10
print(x)
```

When we write:

x = 10

Python:
1. Creates an integer object 10
2. Binds the name `x` to that object

So a variable is just a label pointing to an object.

## 2. Objects in Python

Everything in Python is an object:
- int
- float
- str
- list
- dict
- function

Each object has:
- Identity (memory address)
- Type
- Value

```python
x = 10
print("Value:", x)
print("Type:", type(x))
print("Identity:", id(x))
```

## 3. Rebinding

Rebinding means changing what a variable refers to.

It does NOT modify the object.
It changes the reference.

```python
x = 5
print("Before:", x, id(x))

x = 10
print("After:", x, id(x))
```

## 4. Mutation

Mutation means modifying the internal state of an object.

This happens only for mutable objects.

```python
a = [1, 2]
print("Before:", a, id(a))

a.append(3)
print("After:", a, id(a))
```

## 5. Mutable vs Immutable

Mutable:
- list
- dict
- set

Immutable:
- int
- float
- str
- tuple

## 6. Identity vs Equality

== compares values.
is compares memory identity.

```python
a = [1, 2]
b = [1, 2]

print("a == b:", a == b)
print("a is b:", a is b)
```
